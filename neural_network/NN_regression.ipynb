# Neural Network Evaluation

### Regression

```python
X = np.array(np.linspace(-3, 3, 50)).reshape(50,1)
Y_square = X**2
Y_sin = np.sin(X)
Y_abs = np.abs(X)

Y_true = [Y_square, Y_sin, Y_abs]
```

```python
Y_pred = []
f, ax = plt.subplots(3, sharex='col')

for i in range(len(Y_true)):
    model = Model()

    model.add(Dense(input_dim=1, output_dim=3))
    model.add(Activation(func_type='sigmoid'))
    model.add(Dense(output_dim=3))
    model.add(Activation(func_type='sigmoid'))
    model.add(Dense(output_dim=1))

    model.compose(loss = Loss('squared_error'))
    err = model.fit(X, Y_true[i], epoch=1000, alpha = 0.01)
    
    Y_pred.append(model.predict(X))
    
    with plt.style.context('ggplot'):
        ax[i].plot(X, Y_true[i], '.', color = '#3385ff')
        ax[i].hold(True)
        ax[i].plot(X, Y_pred[-1], color = '#ff471a')
```

```python
f, ax = plt.subplots(3, sharex='col')
ax[0].set_title('f(x) = x^2')
ax[1].set_title('f(x) = sin(x)')
ax[2].set_title('f(x) = |x|')

for i in range(len(Y_true)):
    with plt.style.context('ggplot'):
        ax[i].plot(X, Y_true[i], '.', color = '#3385ff')
        ax[i].hold(True)
        ax[i].plot(X, Y_pred[i], color = '#ff471a')
```
